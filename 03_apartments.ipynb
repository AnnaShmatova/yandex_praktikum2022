## Предобработка данных

### Расположение столбцов

Для удобства восприятия данных, переставим столбцы местами более логично.

```python
data = df_1.reindex(columns=[
 'locality_name',
 'rooms',
 'balcony',
 'floor',
 'floors_total',
 'total_area',
 'living_area',
 'kitchen_area',
 'ceiling_height',
 'is_apartment',
 'studio',
 'open_plan',  
 'total_images',
 'last_price',
 'first_day_exposition',
 'days_exposition',
 'cityCenters_nearest',
 'airports_nearest',
 'parks_around3000',
 'parks_nearest',
 'ponds_around3000',
 'ponds_nearest',
])
data
```

### Удаление и замена пропусков

Начнем с логичных пропусков:
- Пропуски в столбце с балконами, заменим на 0. Так как логично предположить, что балкон отсутствует в таких квартирах. 
- Удалим строки с пропусками в столбце с количеством этажей, так как доля таких значений в общем количестве данных незначительна, а для работы с данными данный столбец - количество этажей - нам крайне важен.
- Также удалим строки, где отстуствует населенный пункт, так как доля таких значений тоже незначительна и для анализа мы не сможем использовать эти строки
- Заменим пропущенные значения в столбце is_apartment на False, так как логично предположить, что пропуск является тем, что объект не является апартаментами

```python
data['balcony'] = data['balcony'].fillna(0)
data.dropna(subset=['floors_total'], inplace=True)
data['floors_total'].isna().sum()
```

```python
data.dropna(subset=['locality_name'], inplace=True)
data['locality_name'].isna().sum()
```

```python
data['is_apartment'] = data['is_apartment'].fillna(False)
data['is_apartment'].unique()
```

Переходим к более сложным пропускам, в которых предполагаем связь с другими данными из таблицы. Логично предположить жилая площадь связана с общей площадью. Проверим это предположение. Посчитаем коэффициент Пирсона для всех столбцов по столбцу total_area.

```python
corr = data.corr()['total_area']
corr
```

Как мы видим есть наибольшая связь со столбцом living_area. Проверим ещё на графике эту связь.

```python
data.plot(x='total_area', y='living_area', ylim=(0, 300), xlim=(0, 400), style='o', grid=True,  figsize=(16, 6))
```

Связь очевидна. Чем больше общая площадь, тем больше жилая площадь. На основании вывода выше, заменим пропуски: 
1. Посчитаем медиану по диапазонам площади квартир
2. Заменим пропуски в столбце living_area на медианное значение жилой площади при таком же диапазоне общей площади.

```python
living_50 = data.loc[data['total_area'] <= 50, 'living_area'].median()
print('Медиана жил площ до 50 кв м - ', living_50)

living_50_100 = data.loc[(50 < data['total_area']) & (data['total_area'] <= 100), 'living_area'].median()
print('Медиана жил площ c 50 до 100 кв м - ', living_50_100)

living_100_150 = data.loc[(100 < data['total_area']) & (data['total_area'] <= 150), 'living_area'].median()
print('Медиана жил площ c 100 до 150 кв м - ', living_100_150)

living_150_200 = data.loc[(150 < data['total_area']) & (data['total_area'] <= 200), 'living_area'].median()
print('Медиана жил площ c 150 до 200 кв м - ', living_150_200)

living_200_250 = data.loc[(200 < data['total_area']) & (data['total_area'] <= 250), 'living_area'].median()
print('Медиана жил площ c 200 до 250 кв м - ', living_200_250)

living_250_300 = data.loc[(250 < data['total_area']) & (data['total_area'] <= 300), 'living_area'].median()
print('Медиана жил площ c 250 до 300 кв м - ', living_250_300)

living_300_350 = data.loc[(300 < data['total_area']) & (data['total_area'] <= 350), 'living_area'].median()
print('Медиана жил площ c 300 до 350 кв м - ', living_300_350)
```

```python
for index in data['total_area']:
        if index <= 50:
            data.loc[data['living_area'].isna(), 'living_area'] = living_50
        if 50 < index <= 100:
            data.loc[data['living_area'].isna(), 'living_area'] = living_50_100
        if 100 < index <= 150:
            data.loc[data['living_area'].isna(), 'living_area'] = living_100_150
        if 150 < index <= 200:
            data.loc[data['living_area'].isna(), 'living_area'] = living_150_200
        if 200 < index <= 250:
            data.loc[data['living_area'].isna(), 'living_area'] = living_200_250
        if 250 < index <= 300:
            data.loc[data['living_area'].isna(), 'living_area'] = living_250_300
        if 300 < index <= 350:
            data.loc[data['living_area'].isna(), 'living_area'] = living_300_350
data['living_area'].isna().sum()
```

Заполним пропуски kitchen_area по тому же принципу, что и в жилой площади. Коэффициент Пирсона у площади кухни по столбцу total_area тоже был высоким.

```python
kitchen_50 = data.loc[data['total_area'] <= 50, 'kitchen_area'].median()
print('Медиана площ кухни до 50 кв м - ', kitchen_50)

kitchen_50_100 = data.loc[(50 < data['total_area']) & (data['total_area'] <= 100), 'kitchen_area'].median()
print('Медиана площ кухни c 50 до 100 кв м - ', kitchen_50_100)

kitchen_100_150 = data.loc[(100 < data['total_area']) & (data['total_area'] <= 150), 'kitchen_area'].median()
print('Медиана площ кухни c 100 до 150 кв м - ', kitchen_100_150)

kitchen_150_200 = data.loc[(150 < data['total_area']) & (data['total_area'] <= 200), 'kitchen_area'].median()
print('Медиана площ кухни c 150 до 200 кв м - ', kitchen_150_200)

kitchen_200_250 = data.loc[(200 < data['total_area']) & (data['total_area'] <= 250), 'kitchen_area'].median()
print('Медиана площ кухни c 200 до 250 кв м - ', kitchen_200_250)

kitchen_250_300 = data.loc[(250 < data['total_area']) & (data['total_area'] <= 300), 'kitchen_area'].median()
print('Медиана площ кухни c 250 до 300 кв м - ', kitchen_250_300)

kitchen_300_350 = data.loc[(300 < data['total_area']) & (data['total_area'] <= 350), 'kitchen_area'].median()
print('Медиана площ кухни c 300 до 350 кв м - ', kitchen_300_350)
```

```python
for index in data['total_area']:
        if index <= 50:
            data.loc[data['kitchen_area'].isna(), 'kitchen_area'] = kitchen_50
        if 50 < index <= 100:
            data.loc[data['kitchen_area'].isna(), 'kitchen_area'] = kitchen_50_100
        if 100 < index <= 150:
            data.loc[data['kitchen_area'].isna(), 'kitchen_area'] = kitchen_100_150
        if 150 < index <= 200:
            data.loc[data['kitchen_area'].isna(), 'kitchen_area'] = kitchen_150_200
        if 200 < index <= 250:
            data.loc[data['kitchen_area'].isna(), 'kitchen_area'] = kitchen_200_250
        if 250 < index <= 300:
            data.loc[data['kitchen_area'].isna(), 'kitchen_area'] = kitchen_250_300
        if 300 < index <= 350:
            data.loc[data['kitchen_area'].isna(), 'kitchen_area'] = kitchen_300_350
data['kitchen_area'].isna().sum()
```

```python
print('Средняя площадь кухни', data['kitchen_area'].median())
print('Средняя жилая площадь', data['living_area'].median())
print('Средняя общая площадь', data['total_area'].median())
```

```python
#проверим не превышает ли общую площадь сумма площади кухни и жилой площади 
print('Сумма средней площади кухни и жилой площади', data['kitchen_area'].median()+data['living_area'].median())
print('Средняя общая площадь', data['total_area'].median())
```

Теперь разберемся с высотой потолков. Есть предположение, что высота потолков напрямую связана с этажностью зданий. Проверим наглядно эту связь на графике.

```python
data.plot(x='floors_total', y='ceiling_height', ylim=(1.5, 8), xlim=(0, 30), style='o', grid=True,  figsize=(16, 6))
```

Как мы видим чем меньше этажей, тем выше потолки. Заменим пропуски в столбце с высотой потолка на медианные значения в зависимости от этажности дома.

```python
for floor in data['floors_total'].unique():
    median = data.loc[data['floors_total'] == floor, 'ceiling_height'].median()
    data.loc[data['ceiling_height'].isna(),'ceiling_height'] = median
data['ceiling_height'].isna().sum()
```

Остался столбец days_exposition. Логично предположить, что на количество дней размещения объявления больше всего вляет населенный пункт. В Санкт-Петербурге медианное значение по столбцу days_exposition будет выше чем в поселке. Проверим это. Для начала приведем к нижнему регистру названия населенных пунктов и заменим ё на е.

```python
#приведем к нижнему регистру названия населенных пунктов
data['locality_name'] = data['locality_name'].str.lower()
data['locality_name'] = data['locality_name'].str.replace('ё', 'е')
with pd.option_context('display.max_rows', None):
    display(data['locality_name'].value_counts())
```

Посмотрим медианное значение в сантк-петербурге, поселке парголово и приозерске.

```python
print(data.loc[(data['locality_name'] == 'санкт-петербург'), 'days_exposition'].median())  	
print(data.loc[(data['locality_name'] == 'поселок парголово'), 'days_exposition'].median())
print(data.loc[(data['locality_name'] == 'приозерск'), 'days_exposition'].median())
```

Предположение подтвердилось. Срок продажи квартир в разных населенных пунктах отличается. Заменим пропуски в days_exposition на медианные значения по населенным пунктам. 

```python
for city in data['locality_name'].unique():
    median = data.loc[data['locality_name'] == city, 'days_exposition'].median()
    data.loc[(data['days_exposition'].isna()) & (data['locality_name'] == city), 'days_exposition'] = median
data['days_exposition'].isna().sum()
```

```python
med = data['days_exposition'].median()
data['days_exposition'] = data['days_exposition'].fillna(med)
```

```python
data.isna().mean()
```

Большое количество пропусков связано скорее всего с неточной настройкой алгоритма - дальше определенного расстояния алгоритм считал, что рядом ничего нет и проставлял пропуск. Данные пропуски удалять не будем, будем воспринимать пропуски как отстуствие этих мест рядом с объектом недвижимости. НО! ### Изменение типов данных

```python
#заменим тип float на int в столбцах где используются целочисленные значения - это удобнее для дальнейшего анализа
data['balcony'] = data['balcony'].astype('int')
data['floors_total'] = data['floors_total'].astype('int')
data['last_price'] = data['last_price'].astype('int')
data['days_exposition'] = data['days_exposition'].astype('int')
# изменим тип данных столбца с первым днем публикации
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
data.info() 
```
